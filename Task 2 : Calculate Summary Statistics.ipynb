# **Task 2 : Calculate Summary Statistics**

```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

##**Importing the Datasets**

```python
#Here the data has been split into two groups in Titanic Dataset : training set(train.csv) and test set(test.csv)
#First Loading Training Set

train_data = pd.read_csv("train.csv")
train_data
```

## **1. Summary Statistics for Training Set**

```python
#info() method prints the information about the DataFrame

train_data.info()
```

```python
#Summary Statistics of Training Set
#describe() method returns description of the data in the DataFrame

train_data.describe()
```

```python
train_data = train_data.select_dtypes(include=['number'])
```

```python
#count() method counts the number of non-null values in each column by default

count = train_data.count()
train_count = pd.DataFrame({'train_count' : count})
train_count
```

```python
#mean() calculates the average of the given dataset

mean = train_data.mean()
train_mean = pd.DataFrame({'train_mean' : mean})
train_mean
```

```python
#median() calculates the middle value of the given dataset

median = train_data.median()
train_median = pd.DataFrame({'train_median' : median})
train_median
```

```python
#mode() returns the value that appears most frequently in the dataset

mode = train_data.mode().iloc[0]
train_mode = pd.DataFrame({'train_mode' : mode})
train_mode
```

```python
#min() returns the items with lowest value

minimum = train_data.min()
train_min = pd.DataFrame({'train_min' : minimum})
train_min
```

```python
#max() returns the item with highest value

maximum = train_data.max()
train_max = pd.DataFrame({'train_max' : maximum})
train_max
```

```python
#std() calculates the standard deviation

std_dev = train_data.std()
train_StandardDeviation = pd.DataFrame({'train_StandardDeviation' : std_dev})
train_StandardDeviation
```

```python
#range is the difference between maximum and minimum values in a dataset

range = train_data.max() - train_data.min()
train_range = pd.DataFrame({'train_range' : range})
train_range
```

```python
#Variance gives variability of the distribution with respect to mean

variance = train_data.var()
train_Variance= pd.DataFrame({'train_Variance' : variance})
train_Variance
```

```python
#Measures of central tendency i.e Mean, Median and Mode for Training Set

mean = train_data.mean()
median = train_data.median()

#Mode may have multiple values, so the first mode from the dataset can returned by .mode() using .iloc[0]

mode = train_data.mode().iloc[0]

#Measures of dispersion i.e Range, Standard Deviation and Variance

data_range = train_data.max() - train_data.min()
std_dev = train_data.std()
variance = train_data.var()

#displaying the Summary Statistics

train_Summary_Statistics = pd.DataFrame({'Mean': mean, 'Median': median, 'Mode': mode, 'Range': data_range, 'Standard Deviation' : std_dev, 'Variance': variance})
print("The Summary Statistics of Training Set is:\n")
print(train_Summary_Statistics)
```
