# **Task 3 : Visualization Using Histogram**

```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## **Importing the Dataset**

```python
#Loading the iris dataset

iris_data = pd.read_csv("iris.csv")
iris_data
```

```python
#dtypes returns the data type of each column in the DataFrame
iris_data.dtypes
```

## **1. Histogram of Sepal Length**

```python
#Plotting histogram of sepal length
#Towers or bars of a histogram are called bins. The height of each bin shows how many values from that data fall into that range
#alpha parameter determines the transparency of the plotted elements. It's a value between 0 and 1, where 0 is invisible and 1 is fully visible.

plt.figure(figsize=(9, 7))
plt.hist(iris_data["SepalLengthCm"], bins=10, color='lightgreen', edgecolor='black', alpha=0.7)
plt.xlabel('Sepal Length in cm')
plt.ylabel('Frequency')
plt.title('Histogram of Sepal Length in Iris Dataset')
plt.show()
```

## **2. Histogram of Sepal Width**

```python
#Plotting Histogram of Sepal Width

plt.figure(figsize=(9, 7))
plt.hist(iris_data["SepalWidthCm"], bins=10, color='crimson', edgecolor='black', alpha=0.7)
plt.xlabel('Sepal Width in cm')
plt.ylabel('Frequency')
plt.title('Histogram of Sepal Width in Iris Dataset')
plt.show()
```

## **3. Histogram of Petal Length**

```python
#Plotting Histogram of Petal Length

plt.figure(figsize=(9, 7))
plt.hist(iris_data["PetalLengthCm"], bins=10, color='turquoise', edgecolor='black', alpha=0.7)
plt.xlabel('Petal Length in cm')
plt.ylabel('Frequency')
plt.title('Histogram of Petal Length in Iris Dataset')
plt.show()
```

## **4. Histogram of Petal Width**

```python
#Plotting Histogram of Petal Width

plt.figure(figsize=(9, 7))
plt.hist(iris_data["PetalWidthCm"], bins=10, color='violet', edgecolor='black', alpha=0.7)
plt.xlabel('Petal Width in cm')
plt.ylabel('Frequency')
plt.title('Histogram of Petal Width in Iris Dataset')
plt.show()
```

```python
#Counting the occurrences of each species

species_counts = iris_data['Species'].value_counts()

#Plotting the distribution of species

plt.figure(figsize=(9, 7))
species_counts.plot(kind='bar', color='orange')
plt.title('Distribution of Species in Iris Dataset')
plt.xlabel('Species')
plt.ylabel('Count')
# Rotating x-axis labels for better readability
plt.xticks(rotation=360)
plt.show()
```
